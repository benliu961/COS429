```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
similarities = "C:/Users/Benjamin Liu/Downloads/caption_similarity_GIT.csv"
data = pd.read_csv(similarities)

sim = data['similarity'].to_numpy()
csim = data['caption similarity'].to_numpy()

plt.scatter(sim, csim)
plt.xlabel('Similarity')
plt.ylabel('Caption Similarity')
plt.title('Similarity vs Caption Similarity')
plt.show()
```
